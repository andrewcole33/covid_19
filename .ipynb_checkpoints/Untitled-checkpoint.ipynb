```python
full_df = pd.read_csv('time-series-19-covid-combined_csv.csv')
full_df.info()
```

```python
columns = ['date', 'country', 'state', 'lat', 'long', 'confirmed', 'recovered', 'deaths']
full_df.columns = columns
```

```python
full_df.date = pd.to_datetime(full_df.date)
full_df.sort_values('date', ascending = True, inplace = True)
full_df = full_df.set_index('date')
full_df.recovered = full_df.recovered.fillna(0)
full_df['conf_pct_change'] = full_df['confirmed'].pct_change(periods = 1).fillna(0)
full_df['rec_pct_change'] = full_df['recovered'].pct_change(periods = 1).fillna(0)
full_df['death_pct_change'] = full_df['deaths'].pct_change(periods = 1).fillna(0)
full_df
```
